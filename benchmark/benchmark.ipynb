## Small Number of Qubits.

```python
# ----------------------------------------------------
# --- Configuration ---
# ----------------------------------------------------

# QFT Configuration
QFT_QUBIT_RANGE = list(range(2, 10))  # 2 to 7 qubits (max 128 dimensions)
QFT_NUM_SHOTS = 100

# QPE Configuration (Note: QPE uses N estimation qubits + 1 target qubit)
QPE_QUBIT_RANGE = list(range(1, 10)) # 1 to 6 estimation qubits (2 to 7 total qubits)
QPE_NUM_SHOTS = 100

# GROVERS Config
GROVER_QUBIT_RANGE = list(range(2, 8)) # (2 to 7 total qubits)
GROVER_NUM_SHOTS = 100

# The list of methods to actually run in the benchmark
METHODS_TO_BENCHMARK = ['dense', 'dense_opt', 'sparse', 'sparse_opt', 'hybrid', 'qiskit_aer', 'qiskit_opt']
```

```python
# ----------------------------------------------------
# --- QFT Benchmark Run ---
# ----------------------------------------------------

print("\n" + "="*50)
print("Starting QFT Benchmarking")
print("="*50)

# Run the benchmark
qft_benchmark_results = benchmark_qft(QFT_QUBIT_RANGE, QFT_NUM_SHOTS, METHODS_TO_BENCHMARK)

# Plot the results
plot_benchmarks(qft_benchmark_results, QFT_NUM_SHOTS, circuit_title='QFT')
```

```python
# ----------------------------------------------------
# --- QPE Benchmark Run ---
# ----------------------------------------------------

print("\n" + "="*50)
print("Starting QPE Benchmarking")
print("="*50)

# Run the benchmark
qpe_benchmark_results = benchmark_qpe(QPE_QUBIT_RANGE, QPE_NUM_SHOTS, METHODS_TO_BENCHMARK)

# Plot the results
plot_benchmarks(qpe_benchmark_results, QPE_NUM_SHOTS, circuit_title='QPE')
```

```python
# ----------------------------------------------------
# --- Grovers Benchmark Run ---
# ----------------------------------------------------

print("\n" + "="*50)
print("Starting QPE Benchmarking")
print("="*50)

# Run the benchmark
grover_benchmark_results = benchmark_grover(GROVER_QUBIT_RANGE, GROVER_NUM_SHOTS, METHODS_TO_BENCHMARK)

# Plot the results
plot_benchmarks(grover_benchmark_results, GROVER_NUM_SHOTS, circuit_title='Grover')
```
